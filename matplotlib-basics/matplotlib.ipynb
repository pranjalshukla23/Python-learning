```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(50) * 100
Y_data = np.random.random(50) * 100
```

## 1. Scatter plot

```python
plt.scatter(X_data, Y_data, c = "#000", marker = "*", s = 150, alpha = 0.3)
plt.show()
```

## 2. Line plot

```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83, 80, 82, 82, 83, 81, 80, 79]
print(years)
print(weights)
```

```python
plt.plot(years, weights, c = "green", lw = 3, linestyle = "--")
plt.show()
```

## 3. Bar chart

```python
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 1, 45]
```

```python
plt.bar(x, y,  color = "red", align = "edge", width = 0.1, edgecolor = "green", lw = 6)
plt.show()
```

## 4. Histogram

```python
ages = np.random.normal(20, 1.5, 1000)
print(ages)
```

```python
plt.hist(ages, bins = [ages.min(), 18, 21, ages.max()], cumulative=True)
plt.show()
```

## 5. Pie chart

```python
langs = ["Python", "C++", "Java", "C#", "Go"]
votes = [50, 24, 14, 6, 17]
explodes = [0, 0, 0, 0.2, 0]
```

```python
plt.pie(votes, labels=langs, explode = explodes, autopct = "%.2f%%", pctdistance = 1.5, startangle = 90)
plt.show()
```

## 6. Box plot

```python
heights = np.random.normal(172, 8, 300)
```

```python
plt.boxplot(heights)
plt.show()
```

## 7. Plot customization

```python
years = [2014, 2015, 2016, 2017, 2018, 2019 , 2020, 2021]
income = [55, 56, 62, 61, 72, 72, 73, 75]
income_ticks = list(range(50, 81, 2))
```

```python
plt.plot(years, income)
plt.title("Income of John (in USD)", fontsize = 30, fontname = "Arial")
plt.xlabel("Years")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f"{x}k" for x in income_ticks])
plt.show()
```

## 10. Multiple figures

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

## 10. Subplots

```python
x = np.arange(100)

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine wave")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cosine wave")

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("Random function")

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Log function")
axs[1,1].set_xlabel("TEST")

fig.suptitle("Four Plots")

plt.show()
```

# 12. 3D Plot

```python
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```
