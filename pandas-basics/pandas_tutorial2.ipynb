## Apply function

```python
# apply function helps you to apply a custom function to every value of a column and return a new series
bios_new["height_category"] = bios_new["height_cm"].apply(lambda x: "Short" if x < 165 else "Average" if x  < 185 else "Tall")
```

```python
bios_new
```

```python
# custom function which takes a row as input
def categorize_athlete(row):
    # accessing a specific column value
    if (row["height_cm"] < 175) and (row["weight_kg"] < 70):
        return "Lightweight"
    elif row["height_cm"] < 185 or row["weight_kg"] <= 80:
        return "Middleweight"
    else:
        return "Heavyweight"
```

```python
# applying a custom function to the entire dataframe
bios["Category"] = bios.apply(categorize_athlete, axis=1)
```

## Handling null values

```python
coffee_df = pd.read_csv("./warmup_data/coffee.csv")
```

### 1. fillna()
- replaces missing values with specific values

```python
# replace missing values with a specific value
coffee_df.fillna(coffee_df["Units Sold"].mean(), inplace=True)
```

### 2. dropna()
- drops rows with missing values
- you can also specify the rows to drop by checking null values in a specific column using subset parameter

```python
# drops all the rows with missing values in Units sold column
coffee_df.dropna(subset=["Units Sold"], inplace=True)
```

```python
# get the number of missing values in Units Sold column
coffee_df["Units Sold"].isna().value_counts()
```

### 3. Pivot table

The pivot function in pandas reshapes your data by rearranging rows into columns based on specified keys. This can be helpful when you need to summarize or reorganize your data for easier analysis. Here’s a simple explanation of its main parameters:

index: Specifies the column(s) you want to use as the new row labels (index) of the pivoted table. Each unique value in this column will become a row in the output.

columns: Specifies the column whose unique values you want to turn into new columns in the pivoted table.

values: Defines the column whose values you want to display in the cells of the pivoted table.

```python
pivot = coffee_df.pivot(columns="Coffee Type", index="Day", values="Units Sold")
```
