```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

## 1. Line plot

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event", style = "event")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="event", style = "event", markers = True)
plt.show()
```

## 2. Barplot

```python
import pandas as pd
```

```python
pokemon = pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
sns.barplot(x="is_legendary", y="speed", data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary", y="weight_kg", data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary", y="speed", hue="generation", data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary", y="speed", hue="generation", palette="Blues_d", data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary", y="speed", hue="generation", color="orange", data=pokemon)
plt.show()
```

## 3. Scatter plot

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", data=iris)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", color="Species", data=iris)
plt.show()
```

```python
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", hue="PetalLengthCm", color="PetalLengthCm", data=iris)
plt.show()
```

## 4. Histogram / Distplot

```python
diamonds = pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
sns.distplot(diamonds["price"])
plt.show()
```

```python
sns.distplot(diamonds["price"], hist=False)
plt.show()
```

```python
sns.distplot(diamonds["price"], color="green")
plt.show()
```

```python
sns.distplot(diamonds["price"], color="green", bins = 10, kde = False)
plt.show()
```

```python
sns.distplot(diamonds["price"], color="green", vertical = True)
plt.show()
```

## 5. Jointplot

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", data=iris)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", color="red", data=iris)
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="PetalLengthCm", color="red", data=iris, kind="reg")
plt.show()
```

## 6. Boxplot

```python
churn = pd.read_csv("churn.csv")
churn.head()
```

```python
sns.boxplot(x="Churn", y="tenure", data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService", y="MonthlyCharges", data=churn)
plt.show()
```

```python
sns.boxplot(x="Contract", y="tenure", data=churn, palette="Set1", linewidth=3)
plt.show()
```

```python
sns.boxplot(x="Contract", y="tenure", data=churn, hue="PaymentMethod")
plt.show()
```
